#  Same Tree

## Problem Statement
Given the roots of two binary trees `p` and `q`, write a function to check if they are the same or not.

Two binary trees are considered the same if they are structurally identical, and the nodes have the same value.

## Examples
```
Input: p = [1,2,3], q = [1,2,3]
Output: true

Input: p = [1,2], q = [1,null,2]
Output: false

Input: p = [1,2,1], q = [1,1,2]
Output: false
```

````python
from collections import deque

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def is_same_tree_recursive(p, q):
    """
    Recursive Approach
    Time Complexity: O(min(m, n))
    Space Complexity: O(min(m, n)) - recursion stack
    """
    # Base cases
    if not p and not q:
        return True
    if not p or not q:
        return False
    
    # Check current nodes and recursively check subtrees
    return (p.val == q.val and 
            is_same_tree_recursive(p.left, q.left) and 
            is_same_tree_recursive(p.right, q.right))

def is_same_tree_iterative(p, q):
    """
    Iterative Approach using Stack
    Time Complexity: O(min(m, n))
    Space Complexity: O(min(m, n))
    """
    stack = [(p, q)]
    
    while stack:
        node1, node2 = stack.pop()
        
        if not node1 and not node2:
            continue
        if not node1 or not node2:
            return False
        if node1.val != node2.val:
            return False
        
        stack.append((node1.left, node2.left))
        stack.append((node1.right, node2.right))
    
    return True

def is_same_tree_bfs(p, q):
    """
    BFS Approach using Queue
    Time Complexity: O(min(m, n))
    Space Complexity: O(min(m, n))
    """
    queue = deque([(p, q)])
    
    while queue:
        node1, node2 = queue.popleft()
        
        if not node1 and not node2:
            continue
        if not node1 or not node2:
            return False
        if node1.val != node2.val:
            return False
        
        queue.append((node1.left, node2.left))
        queue.append((node1.right, node2.right))
    
    return True

def build_tree(arr):
    if not arr or arr[0] is None:
        return None
    
    root = TreeNode(arr[0])
    queue = deque([root])
    i = 1
    
    while queue and i < len(arr):
        node = queue.popleft()
        
        if i < len(arr) and arr[i] is not None:
            node.left = TreeNode(arr[i])
            queue.append(node.left)
        i += 1
        
        if i < len(arr) and arr[i] is not None:
            node.right = TreeNode(arr[i])
            queue.append(node.right)
        i += 1
    
    return root

# Test cases
test_cases = [
    ([1, 2, 3], [1, 2, 3]),          # Same trees
    ([1, 2], [1, None, 2]),          # Different structure
    ([1, 2, 1], [1, 1, 2]),          # Different values
    ([], []),                         # Both empty
    ([1], [1])                        # Single nodes
]

print("🔍 Same Tree:")
for i, (arr1, arr2) in enumerate(test_cases, 1):
    p = build_tree(arr1)
    q = build_tree(arr2)
    
    recursive_result = is_same_tree_recursive(p, q)
    iterative_result = is_same_tree_iterative(p, q)
    bfs_result = is_same_tree_bfs(p, q)
    
    print(f"Test {i}: {arr1} vs {arr2} → {recursive_result}")
````

## 💡 Key Insights

### Recursive Logic
1. **Both None**: Trees are same
2. **One None**: Trees are different
3. **Both exist**: Check value equality and recurse on subtrees

### Early Termination
- Stop as soon as difference is found
- Time complexity is O(min(m, n)) not O(m + n)
- Only traverse until first mismatch

### Three Traversal Methods
1. **Recursive**: Most natural and readable
2. **Iterative DFS**: Uses stack to simulate recursion
3. **BFS**: Compares level by level

## 🎯 Practice Tips
1. Handle None cases carefully in base conditions
2. Short-circuit evaluation with logical operators
3. This pattern extends to "subtree" problems
4. Consider which traversal method fits the problem best
5. Recursive solutions often most intuitive for tree problems
