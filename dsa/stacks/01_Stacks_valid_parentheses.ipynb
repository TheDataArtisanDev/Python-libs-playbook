# 02 Valid Parentheses - Stack Problem

## Problem Statement
Given a string `s` containing just the characters `'('`, `')'`, `'{'`, `'}'`, `'['` and `']'`, determine if the input string is valid.

An input string is valid if:
1. Open brackets must be closed by the same type of brackets
2. Open brackets must be closed in the correct order
3. Every close bracket has a corresponding open bracket of the same type

## Examples
```
Input: s = "()"
Output: true

Input: s = "()[]{}"
Output: true

Input: s = "(]"
Output: false

Input: s = "([)]"
Output: false

Input: s = "{[]}"
Output: true
```

## Constraints
- 1 ‚â§ s.length ‚â§ 10‚Å¥
- s consists of parentheses only '()[]{}'

````python
def is_valid_parentheses(s):
    """
    Stack-based Solution
    Time Complexity: O(n)
    Space Complexity: O(n)
    """
    stack = []
    mapping = {')': '(', '}': '{', ']': '['}
    
    for char in s:
        if char in mapping:  # Closing bracket
            if not stack or stack.pop() != mapping[char]:
                return False
        else:  # Opening bracket
            stack.append(char)
    
    return len(stack) == 0

# Test cases
test_cases = ["()", "()[]{}", "(]", "([)]", "{[]}", "", "(((", ")))", "({[]})"]

print("üîç Valid Parentheses Solutions:")
print("=" * 40)

for i, s in enumerate(test_cases, 1):
    result = is_valid_parentheses(s)
    print(f"Test Case {i}: '{s}' ‚Üí {result}")
````

## üí° Key Insights

### Stack Pattern Recognition
- **LIFO Nature**: Last In, First Out matches bracket pairing
- **Push**: Opening brackets go onto stack
- **Pop**: Closing brackets must match top of stack
- **Empty Stack**: All brackets properly matched

### Algorithm Steps
1. **Opening bracket** ‚Üí Push to stack
2. **Closing bracket** ‚Üí Check if matches stack top
3. **Mismatch or empty stack** ‚Üí Invalid
4. **Final check** ‚Üí Stack should be empty
