```python
import numpy as np
import pandas as pd
```

```python
print("---- Missing Data -----")

# Create a DataFrame with missing values
df = pd.DataFrame({
    'A': [1, 2, np.nan, 4, 5],
    'B': [10, np.nan, 30, 40, 50],
    'C': [np.nan, 200, 300, 400, 500]
})

print("DataFrame with missing values:")
print(df)
print()

# Detect missing values
print("Detect NA values:")
print(df.isna())
print()

print("Detect non-NA values:")
print(df.notna())
print()

# Fill missing values with a value
print("Fill NA with 0:")
print(df.fillna(0))
print()

# Forward fill
print("Forward fill NA values:")
print(df.fillna(method='ffill'))
print()

# Backward fill
print("Backward fill NA values:")
print(df.fillna(method='bfill'))
print()

# Fill missing values with different values for each column
print("Fill NA with different values for each column:")
print(df.fillna({'A': 0, 'B': 20, 'C': 100}))
print()

# Interpolate
print("Interpolate missing values:")
print(df.interpolate())
print()

# Drop rows with NA values
print("Drop rows with any NA values:")
print(df.dropna())
print()

# Drop rows where all values are NA
print("Drop rows where all values are NA:")
print(df.dropna(how='all'))
print()

# Drop rows with NA in specific columns
print("Drop rows with NA in column 'A':")
print(df.dropna(subset=['A']))
print()
```
