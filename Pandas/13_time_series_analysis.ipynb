# <a id='toc1_'></a>[Import Required Libraries](#toc0_)
This cell imports the necessary libraries and sets display options for pandas.

```python
# Import required libraries
import pandas as pd
import numpy as np

# Set display options for better readability
pd.set_option('display.max_rows', 10)
pd.set_option('display.max_columns', 10)
pd.set_option('display.width', 100)
```

# <a id='toc2_'></a>[Creating Time Series Data](#toc0_)
This cell demonstrates how to create time series data using pandas. It includes examples of creating `DatetimeIndex`, `Series`, and `DataFrame` objects with time-based indices.

```python
# 1. Creating Time Series Data
# -------------------------------------------------
# Create a DatetimeIndex with daily frequency
dates = pd.date_range(start='2023-01-01', periods=10, freq='D')
print(f"DatetimeIndex:\n{dates}\n")

# Create a Series with DatetimeIndex as the index
ts = pd.Series(np.random.randn(10), index=dates)
print(f"Time Series:\n{ts}\n")

# Create a DataFrame with DatetimeIndex as the index
df_ts = pd.DataFrame({
    'A': np.random.randn(10),
    'B': np.random.randn(10) + 1
}, index=dates)
print(f"DataFrame with DatetimeIndex:\n{df_ts}\n")

# Create date ranges with different frequencies
weekly = pd.date_range('2023-01-01', periods=5, freq='W')
monthly = pd.date_range('2023-01-01', periods=5, freq='M')
quarterly = pd.date_range('2023-01-01', periods=4, freq='Q')
yearly = pd.date_range('2023-01-01', periods=3, freq='Y')
business_days = pd.date_range('2023-01-01', periods=5, freq='B')
hourly = pd.date_range('2023-01-01', periods=5, freq='H')

# Print all the ranges to understand their structure
print(f"Weekly:\n{weekly}\n")
print(f"Monthly:\n{monthly}\n")
print(f"Quarterly:\n{quarterly}\n")
print(f"Yearly:\n{yearly}\n")
print(f"Business Days:\n{business_days}\n")
print(f"Hourly:\n{hourly}\n")
```

# <a id='toc3_'></a>[Extracting Date and Time Components](#toc0_)
This cell shows how to extract various date and time components (like year, month, day, etc.) from a `DatetimeIndex`.

```python
# 2. Date and Time Components
# -------------------------------------------------
# Create a DatetimeIndex with hourly frequency
datetime_idx = pd.date_range('2023-01-01 12:30:45', periods=5, freq='H')
print(f"DatetimeIndex with time components:\n{datetime_idx}\n")

# Extract date components
year = datetime_idx.year
month = datetime_idx.month
day = datetime_idx.day
dayofweek = datetime_idx.dayofweek  # Monday=0, Sunday=6
dayofyear = datetime_idx.dayofyear
quarter = datetime_idx.quarter
print(f"Year: {year}\nMonth: {month}\nDay: {day}\nDay of Week: {dayofweek}\nDay of Year: {dayofyear}\nQuarter: {quarter}\n")

# Extract time components
hour = datetime_idx.hour
minute = datetime_idx.minute
second = datetime_idx.second
print(f"Hour: {hour}\nMinute: {minute}\nSecond: {second}\n")
```

# <a id='toc4_'></a>[Time Zone Handling](#toc0_)
This cell demonstrates how to handle time zones in pandas, including localizing naive datetime objects and converting between time zones.

```python
# 3. Time Zone Handling
# -------------------------------------------------
# Create a timezone-naive DatetimeIndex
naive_idx = pd.date_range('2023-01-01', periods=5, freq='D')
print(f"Naive DatetimeIndex:\n{naive_idx}\n")

# Localize to a specific timezone (UTC)
utc_idx = naive_idx.tz_localize('UTC')
print(f"UTC Localized:\n{utc_idx}\n")

# Convert to a different timezone (US/Eastern)
eastern = utc_idx.tz_convert('US/Eastern')
print(f"Converted to US/Eastern:\n{eastern}\n")

# Create timezone-aware DatetimeIndex directly
aware_idx = pd.date_range('2023-01-01', periods=5, freq='D', tz='US/Pacific')
print(f"Timezone-aware DatetimeIndex:\n{aware_idx}\n")
```

```python
# 4. Resampling
# -------------------------------------------------
# Create a time series
dates = pd.date_range('2023-01-01', periods=100, freq='H')
ts = pd.Series(np.random.randn(100), index=dates)
print(f"Original Time Series:\n{ts}\n")

# Downsampling (reducing frequency)
daily_mean = ts.resample('D').mean()
daily_sum = ts.resample('D').sum()
daily_min = ts.resample('D').min()
daily_max = ts.resample('D').max()
print(f"Daily Mean:\n{daily_mean}\n")
print(f"Daily Sum:\n{daily_sum}\n")
print(f"Daily Min:\n{daily_min}\n")
print(f"Daily Max:\n{daily_max}\n")

# Upsampling (increasing frequency)
daily_ts = ts.resample('D').mean()
hourly_ffill = daily_ts.resample('H').ffill()
hourly_bfill = daily_ts.resample('H').bfill()
hourly_interp = daily_ts.resample('H').interpolate(method='linear')
print(f"Hourly Forward Fill:\n{hourly_ffill}\n")
print(f"Hourly Backward Fill:\n{hourly_bfill}\n")
print(f"Hourly Interpolation:\n{hourly_interp}\n")
```

```python
# 5. Shifting and Lagging
# -------------------------------------------------
# Create a simple time series
dates = pd.date_range('2023-01-01', periods=5, freq='D')
ts = pd.Series(range(1, 6), index=dates)
print(f"Original Time Series:\n{ts}\n")

# Shift values forward (lagging)
ts_lag1 = ts.shift(1)
ts_lag2 = ts.shift(2)
print(f"Lagged by 1:\n{ts_lag1}\n")
print(f"Lagged by 2:\n{ts_lag2}\n")

# Shift values backward (leading)
ts_lead1 = ts.shift(-1)
print(f"Leading by 1:\n{ts_lead1}\n")

# Shift the index (not the values)
ts_day_forward = ts.tshift(freq='1D')
print(f"Index shifted forward by 1 day:\n{ts_day_forward}\n")

# Calculate percent change
pct_change = ts.pct_change()
print(f"Percent Change:\n{pct_change}\n")

# Calculate difference
diff = ts.diff()
print(f"Difference:\n{diff}\n")
```

```python
# 6. Rolling and Expanding Windows
# -------------------------------------------------
# Create a longer time series
dates = pd.date_range('2023-01-01', periods=10, freq='D')
ts = pd.Series(range(1, 11), index=dates)
print(f"Original Time Series:\n{ts}\n")

# Rolling windows
rolling_mean = ts.rolling(window=3).mean()
rolling_sum = ts.rolling(window=3).sum()
rolling_std = ts.rolling(window=3).std()
print(f"Rolling Mean:\n{rolling_mean}\n")
print(f"Rolling Sum:\n{rolling_sum}\n")
print(f"Rolling Std Dev:\n{rolling_std}\n")

# Rolling window with min_periods
rolling_flex = ts.rolling(window=3, min_periods=1).mean()
print(f"Rolling with min_periods=1:\n{rolling_flex}\n")

# Expanding windows
expanding_mean = ts.expanding().mean()
expanding_sum = ts.expanding().sum()
print(f"Expanding Mean:\n{expanding_mean}\n")
print(f"Expanding Sum:\n{expanding_sum}\n")

# Exponentially weighted windows
ewm_mean = ts.ewm(alpha=0.3).mean()
print(f"Exponentially Weighted Mean:\n{ewm_mean}\n")
```

```python
# 7. Time Series Visualization
# -------------------------------------------------
dates = pd.date_range('2023-01-01', periods=100, freq='D')
ts1 = pd.Series(np.cumsum(np.random.randn(100)), index=dates)
ts2 = pd.Series(np.cumsum(np.random.randn(100) + 0.5), index=dates)
print(f"Time Series 1:\n{ts1}\n")
print(f"Time Series 2:\n{ts2}\n")

df_ts = pd.DataFrame({'Series1': ts1, 'Series2': ts2})
print(f"DataFrame for Visualization:\n{df_ts}\n")

df_ts.plot(title='Time Series Plot', figsize=(10, 5))
```

```python
# 8. Periods and Frequency Conversion
# -------------------------------------------------
# Create a PeriodIndex
periods = pd.period_range(start='2023-01', periods=5, freq='M')
print(f"PeriodIndex:\n{periods}\n")

# Create a Series with PeriodIndex
period_series = pd.Series(range(1, 6), index=periods)
print(f"Series with PeriodIndex:\n{period_series}\n")

# Convert Period to Timestamp
periods_to_timestamp = periods.to_timestamp()
periods_to_timestamp_end = periods.to_timestamp(how='end')
print(f"Periods to Timestamp (start):\n{periods_to_timestamp}\n")
print(f"Periods to Timestamp (end):\n{periods_to_timestamp_end}\n")

# Convert Timestamp to Period
dates = pd.date_range('2023-01-01', periods=5, freq='D')
dates_to_daily_periods = dates.to_period('D')
dates_to_monthly_periods = dates.to_period('M')
print(f"Dates to Daily Periods:\n{dates_to_daily_periods}\n")
print(f"Dates to Monthly Periods:\n{dates_to_monthly_periods}\n")

# Frequency conversion
monthly_periods = pd.period_range(start='2023-01', periods=3, freq='M')
to_quarterly = monthly_periods.asfreq('Q')
to_weekly_start = monthly_periods.asfreq('W', how='start')
to_weekly_end = monthly_periods.asfreq('W', how='end')
print(f"Monthly Periods:\n{monthly_periods}\n")
print(f"To Quarterly:\n{to_quarterly}\n")
print(f"To Weekly (start):\n{to_weekly_start}\n")
print(f"To Weekly (end):\n{to_weekly_end}\n")
```

```python
# 9. Time Deltas
# -------------------------------------------------
# Create a TimedeltaIndex
deltas = pd.timedelta_range(start='1 day', periods=5, freq='D')
print(f"TimedeltaIndex:\n{deltas}\n")

# Create a Series with TimedeltaIndex
delta_series = pd.Series(range(1, 6), index=deltas)
print(f"Series with TimedeltaIndex:\n{delta_series}\n")

# Arithmetic with dates
start_date = pd.Timestamp('2023-01-01')
date_plus_5days = start_date + pd.Timedelta(days=5)
date_plus_deltas = [start_date + delta for delta in deltas]
print(f"Start Date + 5 Days:\n{date_plus_5days}\n")
print(f"Start Date + Deltas:\n{date_plus_deltas}\n")

# TimedeltaIndex components
days = deltas.days
seconds = deltas.seconds
microseconds = deltas.microseconds
total_seconds = deltas.total_seconds()
print(f"Days: {days}\nSeconds: {seconds}\nMicroseconds: {microseconds}\nTotal Seconds: {total_seconds}\n")
```
